```python

# yaml 파일을 학습이 가능하도록 경로 설정.
# key-value 데이터인 dict 데이터타입으로 data['train'], data['val'], data['nc'], data['names'] 에 넣어주는데,
# 가장 중요한 부분은 데이터 경로 설정임.

import yaml

data = { 'train' : '/content/Violence_Data/train/images/',
         'val' : '/content/Violence_Data/valid/images/',
         'test' : '/content/Violence_Data/test/images',
         'names' : ['0', '1'],
         'nc' : 2 }

with open('/content/Violence_Data/Violence_Data.yaml', 'w') as f:
  yaml.dump(data, f)


with open('/content/Violence_Data/Violence_Data.yaml', 'r') as f:
  aquarium_yaml = yaml.safe_load(f)
  display(aquarium_yaml)
```

```python
import ultralytics

ultralytics.checks()
```

```python
# Load YOLOv8n

from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # load a pretrained YOLOv8n detection model
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(data='/content/Violence_Data/Violence_Data.yaml', epochs=300, patience=30, batch=32, imgsz=416)
```

```python
print(type(model.names), len(model.names))

print(model.names)
```
